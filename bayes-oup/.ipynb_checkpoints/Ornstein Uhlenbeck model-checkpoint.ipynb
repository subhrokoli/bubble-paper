```python
from __future__ import division
import sys
import pystan
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# the Ornstein-Uhlenbeck model
# sufficient statistics are used for efficiency

oup_code = """
data {
    int<lower=0> N;    // number of data points
    real<lower=0> dt;  // sampling interval  
    real x[N];         // positions 
}
transformed data {
    real T1;           // sufficient statistics
    real T2;
    real T3;
    real T4;
    T1 <- 0;
    T2 <- 0;
    T3 <- 0;
    T4 <- x[1]*x[1];
    for (n in 1:N-1){
         T1 <- T1 + x[n+1]*x[n+1];
         T2 <- T2 + x[n+1]*x[n];
         T3 <- T3 + x[n]*x[n];
    }
}
parameters {
    real<lower=0> lambda;  // mean regression rate
    real<lower=0> D;  // diffusion coefficient
} 
transformed parameters {
}
model {
    real a;
    real b;
    real c;
    a <- exp(-lambda*dt);
    b <- 1.0 - exp(-2.0*lambda*dt);
    c <- D/(lambda);
    increment_log_prob( -(N-1)/2 * log(2*3.14159*c*b) - (T1 - 2.0*T2*a + T3*a*a)/(2*b*c) - (log(2*3.14159*c)/2) - (T4/(2*c)));
}    
generated quantities {
}
"""

# model compilation
model = pystan.StanModel(model_code=oup_code)
```

```python
#Now load the lab data
sys.path.append('/home/dipanjan/iiscer-kolkata-traps-bubbles/WorkTree/data/')
df = pd.read_excel('/home/dipanjan/iiscer-kolkata-traps-bubbles/WorkTree/data/water_calibrated_brownian.xlsx',sheetname = 'Sheet1')
x = df['Data(micron)']
N = x.size
dt = 0.5e-3
plt.plot(x);
```

```python
#providing experimental data
trap_data = { 'N':N, 'dt':dt, 'x':x}
```

```python
#model fitting using MCMC
fit_stan = model.sampling(data=trap_data, chains=4, iter=1000)
fit_stan, fit_stan.plot('D'), fit_stan.plot('lambda')
```

```python
#parameters obtianed from Stan
param_stan = fit_stan.extract()
lambda_stan = param_stan['lambda'].mean()
D_stan = param_stan['D'].mean()
```

```python
# model fitting using optimization
fit_optim = model.optimizing(data=trap_data)
fit_optim
```

```python
#parameters obtained from optimizer
lambda_optim = fit_optim['lambda']
D_optim = fit_optim['D']
```

```python
#parameters obtained using analytical expressions
sum_x_squared = sum(x[:-1]*x[:-1]) #sum of x(n)^2 from 1 to N-1
sum_x_pairs = sum(np.array(x[1:])*np.array(x[:-1])) #sum of x(n)*x(n+1)
sum_x_squared_last = sum(x[1:]*x[1:]) #sum of x(n)^2 from 2 to N
lambda_analytical = (1/dt)*np.log(sum_x_squared/sum_x_pairs)
I_two = 1 - np.exp(-2*lambda_analytical*dt)
sum_delta_squared = sum_x_squared_last + (sum_x_squared*(np.exp(-2*lambda_analytical*dt))) - 2*(np.exp(-lambda_analytical*dt))*sum_x_pairs
D_analytical = (lambda_analytical/N)*((x[1]*x[1]) + (sum_delta_squared/I_two))
```

```python
#parameters obtained from stationarity assumption
k_B = 1.38064881313131e-17  # Boltzmann constant (Newton micron/Kelvin)
T = 300                     # temperature (Kelvin)

k_stationary = N*k_B*T/sum(x*x)
```

```python
#Conversion to physical units and comparison of results

k_B = 1.38064881313131e-17  # Boltzmann constant (Newton micron/Kelvin)
T = 300                     # temperature (Kelvin)
print "Trap stiffness from MCMC is", ((lambda_stan/D_stan)*k_B*T)
print "Trap stiffness from optimizer is", ((lambda_optim/D_optim)*k_B*T)
print "Trap stiffness from analysis is", ((lambda_analytical/D_analytical)*k_B*T)
print "Trap stiffness from stationarity is", k_stationary
```

```python
#Plotting the log-posterior
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter

fig = plt.figure()
ax = fig.gca(projection='3d')
lambdap = np.linspace(250, 350, 500)
Dp = np.linspace(0, 1, 500)
lambdap, Dp = np.meshgrid(lambdap, Dp)

I_twop = 1 - np.exp(-2*lambdap*dt)
sum_delta_squaredp = sum_x_squared_last + (sum_x_squared*(np.exp(-2*lambdap*dt))) - 2*(np.exp(-lambdap*dt))*sum_x_pairs
logprob = (((N-1)/2)*np.log(lambdap/2*(np.pi)*Dp*I_twop)) - (lambdap*sum_delta_squaredp/2*Dp*I_two) + (0.5*np.log(lambdap/2*(np.pi)*Dp)) - (0.5*lambdap*(x[0]*x[0])/Dp)

surf = ax.plot_surface(lambdap, Dp, logprob)
plt.show()
```
